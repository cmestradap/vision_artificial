```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import os
from skimage import data
from skimage.filters import threshold_otsu
from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
from skimage.color import label2rgb
from skimage.draw import circle
from skimage.filters import sobel_v,sobel_h, sobel
import skimage 
```

```python
### UTILIDADES PARA LA LECTURA DE IMÁGENES
#Función de lectura rápida de imágenes a color y en escala de grises
def img_read(filename, mode = 'color'):
    if(mode == 'color'):   
        return cv2.cvtColor(cv2.imread(filename),cv2.COLOR_BGR2RGB)
    elif(mode=='grayscale' or mode=='greyscale' or mode == 'gray' or mode == 'grey'):
        return cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    else:
        return None

    #Función para interpolar los puntos entre 0 y 255
def img_scale(img, value_range = [0,255]):
    return (value_range[1] - value_range[0])*(img - np.min(img))/(np.max(img)-np.min(img)) + value_range[0]
    
    #Función para enviar los extremos superior e inferior a 0 y 255 respectivamente en la matriz
def img_trim(img, value_range = [0, 255]):
    res = img.copy()
    res[res > 255] = 255
    res[res < 0] = 0
    return res
```

## Filtro Gamma

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np

def apply_f_on_rgb(img, f, args):
    """
    Función para aplicar una transformación a la imagen
    ----
    Parameters:
        img: imagen a la que le aplicaremos la transformación \n
        f: vble tipo función \n
        args: otros argumentos
    """
    #Crear una matriz de ceros del tamaño de la imagen de entrada
    res = np.zeros(img.shape, np.uint8)
    #Aplicar la transformación f sobre cada canal del espacio de color RGB
    res[:,:,0] = f(img[:,:,0], *args)
    res[:,:,1] = f(img[:,:,1], *args)
    res[:,:,2] = f(img[:,:,2], *args)
    
    return res

def gamma_correction(img, a, gamma):
    
    #Crear copia de la imagen tipo flotante dada la normalización
    img_copy = img.copy().astype(np.float32)/255.0
    #La función corrección gamma es de la forma ax^gamma, donde x es la imagen de entrada
    res_gamma = cv2.pow(img_copy,gamma)
    res = cv2.multiply(res_gamma, a)
    
    #Asegurar que la los datos queden entre 0 y 255 y sean uint8
    res[res<0] = 0
    res = res*255.0
    res[res>255] = 255
    
    res = res.astype(np.uint8)
    
    return res

listaGamma = []
for img in listaColor:
    img = img.astype(np.float32)  
    img = cv2.resize(img, (400,400))
    img = img[100:300,100:300]
    #Definir sigma y el tamaño del kernel
    sigma = 2.5
    k_size = 9
    
    a1 = 3
    gamma1 = 3 #3
    
    img_gamma1 = apply_f_on_rgb(img,gamma_correction,(a1,gamma1))
    
    listaGamma.append(img_gamma1)


lista = []
lista = listaGamma

listaGray = []
for i in range(len(listaGamma)):
    img4_gray = cv2.cvtColor(lista[i], cv2.COLOR_RGB2GRAY)
    listaGray.append(img4_gray)

# bordes = cv2.Canny(img4_gray,100,200)

from skimage import morphology as morph

img_area,thresh2 = cv2.threshold(img4_gray,100,255,cv2.THRESH_BINARY)
thresh2 = thresh2/255
img_area = morph.remove_small_holes(thresh2.astype(bool), area_threshold = 50)
img_area = morph.remove_small_objects(img_area, min_size = 300)

# fig, ax = plt.subplots(2,2,figsize=(20, 20))
# ax[0,0].imshow(img4_gray, cmap="gray")
# ax[0,1].imshow(img_area, cmap="gray")    
# ax[1,0].imshow(bordes, cmap="gray")
#ax[1,1].imshow(lista[11], cmap="gray")
```

## Sobel

```python
listaSobel = []
listaBinaria = []
for img in listaGray:
    img = img.astype(np.float32)  
    img = cv2.resize(img, (400,400))
    img = img[100:300,100:300]
    #La función Sobel recibe: la imagen, el parámetro ddepth, el orden de la derivada que queremos calcular
    #en x, el orden de la deirvada que queremos calcular en y,  el tamaño del kernel.
    gradiente_x = cv2.Sobel(img, ddepth=-5, dx = 1, dy = 0, ksize = 7)
    gradiente_y = cv2.Sobel(img, ddepth=-5, dx = 0, dy = 1, ksize = 7)
    #Calcula la magnitud del sobel, por definición
    magnitud_sobel = np.sqrt(np.power(gradiente_x,2)+np.power(gradiente_y,2))
    #Realizar el escalamiento de la imagen
    magnitud_sobel = img_scale(magnitud_sobel)
    listaSobel.append(magnitud_sobel)
    fig, ax = plt.subplots(1,5,figsize=(20, 20))
    ax[0].set_title("Sobel")
    ax[0].imshow(magnitud_sobel, cmap="gray")
    
    # Aplicamos treshhold

    img_bin,thresh2 = cv2.threshold(magnitud_sobel,50,255,cv2.THRESH_BINARY)
    thresh2 = thresh2/255
    img_bin = morph.remove_small_holes(thresh2.astype(bool), area_threshold = 100)
    img_bin = morph.remove_small_objects(img_bin, min_size = 800)
    listaBinaria.append(img_bin)
    ax[1].set_title("Binaria")
    ax[1].imshow(img_bin, cmap="gray")
    
    radius = 6
    selem = morph.disk(radius)
    cierre = morph.binary_closing(img_bin,selem)
    ax[2].set_title("Cierre")
    ax[2].imshow(cierre, cmap='gray')
    
    menos_areapeq = morph.remove_small_holes(cierre, area_threshold = 10000)
    ax[3].set_title("Agujeros pequeños")
    ax[3].imshow(menos_areapeq, cmap="gray")
    
    menos_areapeq = menos_areapeq.astype(int)
    cierre = cierre.astype(int)
    percution = menos_areapeq - cierre

    ax[4].set_title("Resta")
    ax[4].imshow(percution, cmap='gray')
```
